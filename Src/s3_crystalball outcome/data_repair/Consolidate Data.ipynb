```python
import pandas as pd 
import os, sys 

sys.path.append(os.path.dirname(os.getcwd()))

print(sys.path)
```

```python
tickers = ['adbe', 'ge', 'intu', 'ms', 'vrtx', 'xom']

for ticker in tickers: 
    
    train = pd.read_csv(f'../{ticker}/data/train/OVERALL_prediction_by_model.csv')
    val = pd.read_csv(f'../{ticker}/data/val/OVERALL_prediction_by_model.csv')
    test = pd.read_csv(f'../{ticker}/data/test/OVERALL_prediction_by_model.csv')
    
    consol = pd.concat([train[['Date', 'close']], val[['Date', 'close']], test[['Date', 'close']]], axis = 0)
    consol = consol.set_index('Date')
    
    consol.to_csv(f'{ticker}_retrived_close.csv')
```
